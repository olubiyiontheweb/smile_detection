```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
boy = cv2.imread("eyes.jpeg")
boyc = boy.copy()
```

```python
greyb = cv2.cvtColor(boy, cv2.COLOR_BGR2GRAY)
```

```python
haarface = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
haarsmile = cv2.CascadeClassifier('haarcascade_smile.xml')
```

```python
facesb = haarsmile.detectMultiScale(greyb, 5, 10)
smiles = haarsmile.detectMultiScale(greyb, 2, 10)
#smiles
```

```python
print(str(len(facesb))+" people faces detected")
print(str(len(smiles))+" people smiles detected")
```

```python
#detect faces and only smiles in faces
for faceb in facesb:
    for smile in smiles:
        fxb,fyb,fwb,fhb = faceb
        sxb,syb,swb,shb = smile
        boyface = cv2.rectangle(boyc,(fxb,fyb),((fxb+fwb),(fyb+fhb)), (0,255,0), 5)
        boysmile = cv2.rectangle(boyface,(sxb,syb),((sxb+swb),(syb+shb)), (0,255,0), 5)
```

```python
plt.imshow(boysmile)
```

```python
vick = cv2.imread("vick.jpg")
vick_copy = vick.copy()
```

```python
vgrey = cv2.cvtColor(vick, cv2.COLOR_BGR2GRAY)
vcrgb = cv2.cvtColor(vick_copy, cv2.COLOR_BGR2RGB)
```

```python
faces = haarface.detectMultiScale(vgrey, 1.1, 2)
vsmile = haarsmile.detectMultiScale(vgrey, 1.3, 10)
print(str(len(faces))+" faces detected")
print(str(len(vsmile))+" smiles detected")
```

```python
#detect faces and only smiles in faces
for face in faces:
    for vsmile in vsmile:
        fx,fy,fw,fh = face
        sx,sy,sw,sh = vsmile
        det_face = cv2.rectangle(vcrgb,(fx,fy),(fx+fw,fy+fh),(0,255,0),3)
        det_smile = cv2.rectangle(det_face,(sx,sy),(sx+sw,sy+sh),(0,255,0),3)
```

```python
plt.imshow(det_smile, cmap="gray")
```

```python
students = cv2.imread("face.jpg")
str_copy = students.copy()
body1 = cv2.imread("op.jpeg")
```

```python
grey_stud = cv2.cvtColor(students,cv2.COLOR_BGR2RGB)
plt.imshow(grey_stud)
```

```python
detectedf = haarface.detectMultiScale(grey_stud, scaleFactor=1.2, minNeighbors=2)
```

```python
print(str(len(detectedf))+" approx student faces detected in the class")
```

```python
for face in detectedf:
 # Each face is a rectangle representing
 # the bounding box around the detected face
    x, y, w, h = face
    stu_face = cv2.rectangle(str_copy, (x, y), (x+w, y+h), (0, 255, 205), 3)
```

```python
plt.imshow(stu_face,cmap="gray")
```
